## Messages

- Messages are the fundamental unit of context for models in Langchain. They represent the input and output of models,carrying and metadeta needed to represent the state of a conversation when interacting with a LLM.

- Messages are object that contain:
    - `Role`: Identifies the message type, such as `system`, `user`, or `assistant`.
    - `Content`: Represents the actual content of message (like text, images, documents, etc.).
    - `Metadata`: Optional additional information about the message, such as timestamps, user IDs, or any other relevant data.

- Langchain provides a standard messages type that works across all model providers, ensuring consistent behavior regardless of the model called.

## Messages Prompt

- Alternatively, you can pass list of messages to the model by providing a list of message objects.

- Message types
    - `System message`: Tells the model how to behave and provide context for interaction.
    - `Human message`: Represents user input and interaction with the model.
    - `AI message`: Response generated by model, including text content, tool calls and metadata.
    - `Tool message`: Represents the outputs of tool call.

### System Message
- A system message represent an initial set of instructions that the model's behavior. You can use a system message to set the tone, define the model's role, and establish guidelines for how the model should respond to user inputs. It helps in shaping the overall interaction and ensuring that the model's responses align with the desired context and objectives.

### Human Message
- A human message represent user input and interaction. They can contain text, images, audio, files, and other amount of multimodel content.

### AI Message
- An AI message represent the output invocation. They can include multimodel data, tool calls and provider-specific metadata that you can later access.

### Tool Message
- For models that support tool calling, AI messages can contain tool calls. Tools messages are used to pass the result of a single tool execution back to the model. They can include multimodel data, tool calls and provider-specific metadata that you can later access.

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage(content="You are a data science expert"),
    HumanMessage(content="Explain about RAG applications"),
]

response = model.invoke(messages)
print(response.content)
```

```python
### Detailed info to LLM with System Message
from langchain.messages import SystemMessage, HumanMessage

system_msg = SystemMessage(
    content="""
    You are a senior python developer with expertise in web frameworks.
    Always provide code examples and explain your reasoning.
    Be concise but thorough in your explanations. Avoid unnecessary verbosity.
    """
)

messages = [
    system_msg,
    HumanMessage(content="How do I create a REST API in Python?"),
]

response = model.invoke(messages)
print(response.content)
```

```python
from langchain.messages import AIMessage, SystemMessage, HumanMessage

# Create an AI messages manulally (e.g. for conversation history)
ai_msg = AIMessage(
    content="I'd be happy to help you with that question!"
)

# Add to cnversation history
messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="can you help me?"),
    ai_msg, # Insert as if it come from the model
    HumanMessage(content="Great! What's 5*5?")
]

response = model.invoke(messages)
print(response.content)
```

```python
from json import tool

from langchain.messages import AIMessage
from langchain.messages import ToolMessage

# After a model makes a tool call
# (Here, we demonstrate manually the messages for bravity)
ai_msg = AIMessage(
    content=[],
    tool_call=[{
        "name": "get_weather",
        "arguments": {
            "location": "New York",
            "unit": "celsius"
        },
        "id": "tool_call_123"
    }]
)
```

```python
# Execute tool and create result messages
weather_result = "Sunny, 25°C"
tool_message = ToolMessage(
    content=weather_result,
    tool_call_id="tool_call_123"
)
```

```python
# Continue conversation with the tool result
messages = [
    HumanMessage(content="What's the weather in New York?"),
    ai_msg, # Model's tool call message
    tool_message, # Tool result message
]

response = model.invoke(messages) # Model process the result
print(response.content)
```
