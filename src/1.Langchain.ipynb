## Generative AI Applications Using Agents And Groq Model

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    model="llama-3.1-8b-instant",
    model_provider="groq",
)
```

```python
model
```

```python
## 1st Generative AI Application Using Agents And Groq Model
response = model.invoke("Write a poem about the beauty of nature.")
response
```

```python
print(response.content)
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.7,
)
response = model.invoke("Write a poem about the beauty of coding.")
response
```

```python
print(response.content)
```

### Streaming And Batch
- **Streaming** : 
  - Streaming is a method of processing data in real-time as it is generated or received. It allows for immediate analysis and response to incoming data, making it suitable for applications that require low latency and quick decision-making.
  - In the context of generative AI, streaming can be used to generate content on-the-fly based on user input or other real-time data sources. This can include generating text, images, or other media in response to user interactions.
- **Batch** : 
  - Batch processing involves handling a large volume of data at once, typically in a single operation. This approach is useful for tasks that do not require real-time processing and can benefit from optimized resource utilization.
  - In the context of generative AI, batch processing can be used to generate multiple pieces of content simultaneously, which can be more efficient for certain types of tasks or when dealing with large datasets.

```python
print(model.invoke("Write a 50 word poem on dirty joke on coding.").content)
```

```python
for chunk in model.stream("Write a poem about the beauty of coding."):
    print(chunk.content, end="", flush=True)
```

```python
responses = model.batch(
    [
        "Write a poem about the beauty of coding.",
        "Write a 50 word poem on dirty joke on coding.",
        "Write a poem about the beauty of nature."
    ],
    config = {
        "max_concurrency": 5, # Limit to 5 parallel requests
    }
)

for response in responses:
    print(response)
```
